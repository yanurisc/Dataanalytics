# This is Lab 2 of Exercise 2.B, working with datetime functions. 

```python
import datetime

# The datetime module work with dates and times
```

# Today

```python
today = datetime.datetime.now()

# Datetime.now() grabs the current date AND time from your computer
```

```python
print("Today is " + today.strftime("%A, %B %d, %Y"))

# Strftime lets us format the date however you want
# %A = full day name (Monday), %B = full month name (January), %d = day number, %Y = 4-digit year
```

# My birthday

```python
birthday = datetime.date(1996, 2, 29)

# datetime.date() creates a date using (year, month, day)
```

```python
print("My birthday is " + birthday.strftime("%m/%d/%Y"))

# %m = month as a number, %d = day, %Y = year
# Prints in MM/DD/YYYY format
```

# Next 90 Days

```python
ninety_d = today + datetime.timedelta(days=90)

# timedelta lets you add or subtract time from a date
# You add 90 days to today to get a future date
```

```python
print("90 days from today is " + ninety_d.strftime("%B %d, %Y"))

# Print the date 90 days from now in Month Day, Year format
```

# Time for Dinner

```python
dinner_time = datetime.time(18, 30)

# datetime.time() creates a time using (hour, minute) in 24-hour format
# 18, 30 means 6:30 PM — change this to whatever time you want!
```

```python
print("Let's meet for dinner at " + dinner_time.strftime("%I:%M %p"))

# %I = 12-hour clock, %M = minutes, %p = AM or PM
```
