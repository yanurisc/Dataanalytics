## Lab 1. Working with names – parsing strings

```python
test_name = 'John Doe'
```

```python
def parse_and_display(name):
    for i in range(len(name)):
        if name[i] == ' ':
            first = name[:i]
            last = name[i+1:]
            print(f"Full name:  {name}")
            print(f"First name: {first}")
            print(f"Last name:  {last}")
```

```python
parse_and_display(test_name)
```

```python
test_list = ['Grace Flores', 'JB Oinonen', 'Ken Penn']
# A list of names to test
```

```python
parse_and_display(test_list[0])
# output for Grace Flores
```

```python
parse_and_display(test_list[1])
# output for JB Oinonen
```

```python
parse_and_display(test_list[2])
# output for Ken Penn
```

```python
parse_and_display('Yanuris Campusano')
# output using my name
```

```python
# CHALLENGE: In a new cell at the bottom, create a modified version of your function
# called parse_and_display2() that will work with names made of one, two or
# three words. Output should include middle name if applicable
    
def parse_and_display2(name):
    parts = name.split(' ')
    
    if len(parts) == 1:
        print(f"Full name:   {name}")
        print(f"First name:  {parts[0]}")
        print(f"Last name:   N/A")
    
    elif len(parts) == 2:
        print(f"Full name:   {name}")
        print(f"First name:  {parts[0]}")
        print(f"Last name:   {parts[1]}")
    
    elif len(parts) == 3:
        print(f"Full name:   {name}")
        print(f"First name:  {parts[0]}")
        print(f"Middle name: {parts[1]}")
        print(f"Last name:   {parts[2]}")

# Test all three examples
parse_and_display2('Lorde')
print("---")
parse_and_display2('Billie Eilish')
print("---")
parse_and_display2('Megan Thee Stallion')
```
