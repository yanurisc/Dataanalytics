# Exercise 2.A Using built-in functions
## Lab 2. Parsing strings – part codes

```python
def get_supplier_code(code):
    return code[:code.index(':')]

# Get the supplier name from the part code
# Example: 'ACME:123-L' → returns 'ACME'    
```

```python
def get_product_num(code):
    return code[code.index(':')+1:code.index('-')]

# Get the product number from the part code
# Example: 'ACME:123-L' → returns '123'    
```

```python
def get_size(code):
    return code[code.index('-')+1:]

# Get the size from the part code
# Example: 'ACME:123-L' → returns 'L'    
```

```python
code1 = 'ACME:123-L'
code2 = 'DI:12345-M'
code3 = 'ACE:1-12'

# Three examples codes to test 
```

```python
print(f"For {code1}, the supplier is {get_supplier_code(code1)}. Part number {get_product_num(code1)} is available in size {get_size(code1)}.")
# Test all three functions with code1
```

```python
all_codes = ['ACME:123-L', 'DI:12345-M', 'ACE:1-12']

for code in all_codes:
    print(f"For {code}, the supplier is {get_supplier_code(code)}. Part number {get_product_num(code)} is available in size {get_size(code)}.")
```
