## Lab 3. Employee data
This lab exercise makes updates to information extracted from
an employee database.

```python
hr_list = [
    ('0123', 'HR', 'Rebecca Yang', '69000'),
    ('0121', 'IT', 'Mark Blick', '67000'),
    ('0068', 'IT', 'Kahna Larsen', '112000'),
    ('0234', 'OPS', 'Jim Smith', '54000')
]

# Employee data stored as a list of tuples
# Each tuple contains: employee number, department, name, salary
```

```python
hr_list[1] = ('0121', 'IT', 'Mark Blick-Hawley', '67000')

# Update Mark's last name to Blick-Hawley
# Tuples cannot be changed so we replace the whole tuple for Mark
```

```python
print(hr_list)

# Print the updated list to confirm Mark's name change
```

```python
hr_list[3] = ('0234', 'CS', 'Jim Smith', '60000')

# Update Jim's department from OPS to CS and his salary to 60000
# We replace the whole tuple for Jim with the new information
```

```python
print(hr_list)

# Print the updated list to confirm both changes are saved
# Mark's name and Jim's department and salary should both show the updates
```

```python
for employee in hr_list:
    print(f"{employee[0]} | {employee[1]} | {employee[2]} | {int(employee[3]):,}")

# Display the final list in a neat readable format
# Format: Employee# | DeptCode | Name | NN,NNN    
```
